```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
```

```python
corr = titanic.corr(numeric_only=True)
```

```python
sns.heatmap(data=corr, cmap="YlGnBu", annot=True)
plt.title
plt.show()
```
