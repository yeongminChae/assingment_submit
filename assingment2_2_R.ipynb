```r
library(MASS)
```

```r
tab <- table(Cars93$Manufacturer)
# print(tab)
```

```r
# 파이 차트를 그립니다.
pie(tab, main="Cars93 데이터의 Manufacturer별 빈도수 (파이 차트)")
```

```r
# 막대 그래프를 그립니다.
barplot(tab, main="Cars93 데이터의 Manufacturer별 빈도수 (막대 그래프)",
        xlab="Manufacturer", ylab="빈도수", las=2, col=rainbow(length(tab)), cex.names=0.7)
```

```r
plot351303 <- function(keyword, dataset) {
  if (keyword == "pie") {
    pie(dataset, main="Manufacturer별 빈도수 (파이 차트)")
  } else if (keyword == "bar") {
    barplot(tab, main="Cars93 데이터의 Manufacturer별 빈도수 (막대 그래프)",
        xlab="Manufacturer", ylab="빈도수", las=2, col=rainbow(length(tab)), cex.names=0.7)
  } else {
    cat("오류 발생! : 지원되지 않는 차트 유형입니다. 'pie' 또는 'bar'를 사용해주세요.")
  }
}
```

```r
plot351303("pie", tab)
```

```r
plot351303("bar", tab)
```
